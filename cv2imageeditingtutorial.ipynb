```python
#READING THE IMAGE USING RANDOM
#THIS VERSION MAKES SURE THE LINES HAVE A CERTAIN SIZE
from matplotlib import image
from matplotlib import pyplot as plt
import random  

# to read the image stored in the working directory
data = image.imread('sunset-1373171__340.jpg')

#get height, width,  from (height, width, mode)
width = data.shape[1]
height = data.shape[0]
base_size_x = 50
base_size_y= 20
for r in range(3): # go for 3 iterations
    #we iterate until we find a base size for the line
    while True:
        x1 = random.randint(0, width)
        x2 = random.randint(x1, width)
        if x2 -x1 >= base_size_x:
            break
    while True:
        y1 = random.randint(0, height)
        y2 = random.randint(y1, height)
        if y2-y1 >= base_size_y:
            break
    x = [x1, x2]
    y = [y1, y2]
    plt.plot(x, y, color="white", linewidth=3)
    plt.axis('off')
    plt.imshow(data)
    plt.show()
```
